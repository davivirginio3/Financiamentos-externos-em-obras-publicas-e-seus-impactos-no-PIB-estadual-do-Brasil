```python
# Código completo: leitura, tratamento, agregações e gráficos (SP e RJ)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Caminhos (ajuste se necessário)
pib_path = "PIB dos Municípios - base de dados 2010-2021.xlsx"
fin_path = "dados_2025-02-05.xlsx"

# Leitura (primeira sheet do PIB; planilha 'Dados' do FIN)
pib_sheets = pd.read_excel(pib_path, sheet_name=None)
pib_df = list(pib_sheets.values())[0]

fin_sheets = pd.read_excel(fin_path, sheet_name=None)
fin_df = fin_sheets.get("Dados") if "Dados" in fin_sheets else list(fin_sheets.values())[0]

# --- Preparar PIB ---
pib = pib_df.copy()
pib = pib.rename(columns={
    "Sigla da Unidade da Federação": "UF_sigla",
    "Produto Interno Bruto, \na preços correntes\n(R$ 1.000)": "PIB_valor_1000",
    "Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)": "VAB_Agro_1000",
    "Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)": "VAB_Ind_1000",
    "Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)": "VAB_Serv_1000",
    "Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)": "VAB_AdminSaude_1000"
})

# Filtrar SP e RJ
pib_states = pib[pib["UF_sigla"].isin(["SP","RJ"])].copy()

# Converter para numérico e multiplicar por 1000 (coluna vinha em R$ 1.000)
pib_states["PIB_valor"] = pd.to_numeric(pib_states["PIB_valor_1000"], errors="coerce") * 1000
pib_states["VAB_Agro"] = pd.to_numeric(pib_states["VAB_Agro_1000"], errors="coerce") * 1000
pib_states["VAB_Ind"] = pd.to_numeric(pib_states["VAB_Ind_1000"], errors="coerce") * 1000
pib_states["VAB_Serv"] = pd.to_numeric(pib_states["VAB_Serv_1000"], errors="coerce") * 1000
pib_states["VAB_AdminSaude"] = pd.to_numeric(pib_states["VAB_AdminSaude_1000"], errors="coerce") * 1000

# Agregar por UF e Ano (soma de municípios -> total estadual)
pib_agg = pib_states.groupby(["UF_sigla","Ano"], as_index=False)[["PIB_valor","VAB_Agro","VAB_Ind","VAB_Serv","VAB_AdminSaude"]].sum()
```

```python
# --- Preparar FINANCIAMENTOS ---
fin = fin_df.copy()

# Mapear nmUnidadeFederal para sigla (heurística)
def map_state(name):
    if pd.isna(name):
        return None
    s = name.lower()
    if "são paulo" in s or "sao paulo" in s:
        return "SP"
    if "rio de janeiro" in s:
        return "RJ"
    return None

fin["UF_sigla"] = fin["nmUnidadeFederal"].apply(map_state)

# Converter valor (VlDolar) para numérico
fin["VlDolar_num"] = pd.to_numeric(fin["VlDolar"].astype(str).str.replace(r"[^\d\.-]","", regex=True), errors="coerce")

# Tentar converter datas (priorizar dtRecebimento)
for col in ["dtRecebimento","dtAprovacaoSenado","dtAssinatura","dtAprovDiretoria","dtReuniaoNegociacao"]:
    if col in fin.columns:
        fin[col+"_dt"] = pd.to_datetime(fin[col], dayfirst=True, errors="coerce")

fin["data_evento"] = fin.get("dtRecebimento_dt").fillna(fin.get("dtAssinatura_dt")).fillna(fin.get("dtAprovacaoSenado_dt")).fillna(fin.get("dtAprovDiretoria_dt"))
fin["Ano"] = fin["data_evento"].dt.year

# Filtrar SP e RJ
fin_states = fin[fin["UF_sigla"].isin(["SP","RJ"])].copy()

# Agregar volume por UF e Ano
fin_agg = fin_states.groupby(["UF_sigla","Ano"], as_index=False)["VlDolar_num"].sum()

# Juntar PIB e Financiamentos
merged = pd.merge(pib_agg, fin_agg, on=["UF_sigla","Ano"], how="left").fillna(0)
merged["VlDolar_num_milhoes"] = merged["VlDolar_num"] / 1e6
merged["PIB_milhoes"] = merged["PIB_valor"] / 1e6
```

```python
# 4) Agrupar setores dos financiamentos (mapeamento heurístico)
sector_map = {
    "Agro": ["agric", "agro", "agrícola", "agricultura"],
    "Edu_Saude_Admin": ["educa", "saúde", "saude"],
    "Servicos": ["comércio", "comercio", "serviço","servico","serviços","servicos","turismo","saneamento","desenvolvimento social"],
    "Industria_Infrastr": ["energia","logística","transporte","infra","indústria","industria","mobilidade","infraestrutura","crédito","financiamento"]
}
def map_sector(s):
    if pd.isna(s):
        return "Outros"
    s_low = s.lower()
    for target, keys in sector_map.items():
        for k in keys:
            if k in s_low:
                return target
    return "Outros"

fin_states["sector_group"] = fin_states["nmSetor"].apply(map_sector)
fin_sector_agg = fin_states.groupby(["UF_sigla","Ano","sector_group"], as_index=False)["VlDolar_num"].sum()

# 4b) Plot stacked bar por estado
for uf in ["SP","RJ"]:
    df = fin_sector_agg[fin_sector_agg["UF_sigla"]==uf]
    if df.empty:
        continue
    pivot = df.pivot(index="Ano", columns="sector_group", values="VlDolar_num").fillna(0)/1e6
    pivot.plot(kind="bar", stacked=True, figsize=(10,5))
    plt.title(f"Volume de financiamentos por setor (US$ milhões) - {uf}")
    plt.xlabel("Ano"); plt.ylabel("US$ milhões")
    plt.tight_layout()
    plt.show()


```

```python
# 5) Evolução setorial do PIB (VAB)
for uf in ["SP","RJ"]:
    df = pib_agg[pib_agg["UF_sigla"]==uf].set_index("Ano").sort_index()
    plt.figure(figsize=(8,5))
    for col in ["VAB_Agro","VAB_Ind","VAB_Serv","VAB_AdminSaude"]:
        if df[col].notna().any():
            plt.plot(df.index, df[col]/1e9, marker='o', label=col)
    plt.title(f"Evolução setorial do VAB (R$ bilhões) - {uf}")
    plt.xlabel("Ano"); plt.ylabel("VAB (R$ bilhões)")
    plt.legend(); plt.grid(True); plt.show()

```

```python
# Análises SP e RJ (2010-2021) — código final
import pandas as pd, numpy as np, matplotlib.pyplot as plt, os
from sklearn.linear_model import LinearRegression

# Config
pib_path = "PIB dos Municípios - base de dados 2010-2021.xlsx"
fin_path = "dados_2025-02-05.xlsx"
out_dir = "analise_output"
os.makedirs(out_dir, exist_ok=True)
YEARS_MIN, YEARS_MAX = 2010, 2021
STATES = ["SP","RJ"]

# Load
pib = pd.read_excel(pib_path, sheet_name="PIB dos Municípios")
fin = pd.read_excel(fin_path, sheet_name="Dados")

# Normalize column names
pib.columns = [c.strip() for c in pib.columns]
fin.columns = [c.strip() for c in fin.columns]

# Detect PIB main column
pib_candidates = [c for c in pib.columns if 'Produto Interno Bruto' in c and 'R$' in c]
if not pib_candidates:
    raise RuntimeError("Não encontrei coluna Produto Interno Bruto (R$) no arquivo de PIB.")
pib_col = pib_candidates[0]

# Convert PIB to numeric (note: values are in R$ 1.000 in this dataset)
pib[pib_col] = pd.to_numeric(pib[pib_col], errors='coerce') * 1000

# Optional: detect VAB sector columns (agro/ind/serv/admin)
def find_col_contains(df, terms):
    for c in df.columns:
        low = c.lower()
        if all(t in low for t in terms):
            return c
    return None

vab_agro = find_col_contains(pib, ['agropecuária']) or find_col_contains(pib,['agropecuaria'])
vab_ind = find_col_contains(pib, ['indústria']) or find_col_contains(pib,['industria'])
vab_serv = find_col_contains(pib, ['serviços','exceto']) or find_col_contains(pib,['servicos','exceto'])
vab_admin = find_col_contains(pib, ['administração','saúde']) or find_col_contains(pib,['administracao','saude'])

# Filter to SP & RJ and years 2010-2021
pib_states = pib[pib['Sigla da Unidade da Federação'].isin(STATES) & pib['Ano'].between(YEARS_MIN, YEARS_MAX)].copy()

# Aggregate state-year totals
agg_cols = {pib_col:'PIB_valor'}
if vab_agro: agg_cols[vab_agro] = 'VAB_Agro'
if vab_ind: agg_cols[vab_ind] = 'VAB_Ind'
if vab_serv: agg_cols[vab_serv] = 'VAB_Serv'
if vab_admin: agg_cols[vab_admin] = 'VAB_AdminSaude'

pib_agg = pib_states.groupby(['Sigla da Unidade da Federação','Ano'], as_index=False)[list(agg_cols.keys())].sum().rename(columns=agg_cols)
pib_agg = pib_agg.rename(columns={'Sigla da Unidade da Federação':'UF_sigla'})

# Prepare finance data
def map_state(name):
    if pd.isna(name): return None
    s = str(name).lower()
    if "são paulo" in s or "sao paulo" in s: return "SP"
    if "rio de janeiro" in s: return "RJ"
    return None
fin['UF_sigla'] = fin['nmUnidadeFederal'].apply(map_state)
fin['VlDolar_num'] = pd.to_numeric(fin['VlDolar'].astype(str).str.replace(r"[^\d\.-]","", regex=True), errors='coerce')

# Parse dates and determine year of event
date_cols = ['dtRecebimento','dtAssinatura','dtAprovacaoSenado','dtAprovDiretoria','dtReuniaoNegociacao']
for c in date_cols:
    if c in fin.columns:
        fin[c+'_dt'] = pd.to_datetime(fin[c], dayfirst=True, errors='coerce')
dt_cols = [c+'_dt' for c in date_cols if c+'_dt' in fin.columns]
if dt_cols:
    fin['data_evento'] = fin[dt_cols].bfill(axis=1).iloc[:,0]
    fin['Ano'] = fin['data_evento'].dt.year
else:
    # fallback: if no date column, try any 'Ano' column
    if 'Ano' not in fin.columns:
        raise RuntimeError("Não foi possível identificar coluna de data nos financiamentos.")

# Filter financing to SP/RJ and 2010-2021
fin_states = fin[fin['UF_sigla'].isin(STATES) & fin['Ano'].between(YEARS_MIN, YEARS_MAX)].copy()

# Aggregate finances
fin_agg = fin_states.groupby(['UF_sigla','Ano'], as_index=False).agg(VlDolar_sum=('VlDolar_num','sum'), n_projects=('cdCartaConsulta','count'))

# Merge datasets
merged = pd.merge(pib_agg, fin_agg, on=['UF_sigla','Ano'], how='left').fillna(0)

# Save aggregated csv
pib_agg.to_csv(os.path.join(out_dir,'pib_agg_SP_RJ_2010_2021.csv'), index=False)
fin_agg.to_csv(os.path.join(out_dir,'fin_agg_SP_RJ_2010_2021.csv'), index=False)
merged.to_csv(os.path.join(out_dir,'merged_SP_RJ_2010_2021.csv'), index=False)



```

```python
# 2) Financiamentos timeseries (US$ milhões)
for uf in STATES:
    
    df = merged[merged['UF_sigla']==uf].sort_values('Ano')
    plt.figure(figsize=(8,4))
    plt.plot(df['Ano'], df['VlDolar_sum']/1e6, marker='s', color='C1')
    plt.title(f'Financiamentos ({uf}) — 2010–2021 (US$ milhões)')
    plt.xlabel('Ano'); plt.ylabel('US$ milhões'); plt.grid(True)
    plt.savefig(os.path.join(out_dir,f'Financiamentos_{uf}_2010_2021.png')); plt.show()
```

```python
# 4) Nº projetos x volume por ano (duplo eixo)
import os
import matplotlib.pyplot as plt

for uf in STATES:
    df = merged[merged['UF_sigla'] == uf].sort_values('Ano')

    # Converte valores em dólar -> real
    cotacao = 3.29  # exemplo de cotação média
    df['Valor_BRL'] = df['VlDolar_sum'] * cotacao  

    fig, ax1 = plt.subplots(figsize=(10, 4))

    # Nº de projetos
    ax1.bar(df['Ano'] - 0.15, df['n_projects'], width=0.3, label='n_projects')
    ax1.set_xlabel('Ano')
    ax1.set_ylabel('Nº projetos')

    # Volume em R$
    ax2 = ax1.twinx()
    ax2.bar(df['Ano'] + 0.15, df['Valor_BRL'] / 1e6,  # em milhões
            width=0.3, color='C1', label='Volume (R$ mi)')
    ax2.set_ylabel('Volume (R$ milhões)')

    plt.title(f"Nº projetos e volume ({uf}) — 2010–2021")
    ax1.grid(True, axis='y')

    # Legenda unificada
    fig.legend(loc="upper left")

    # Salvar gráfico
    plt.savefig(os.path.join(out_dir, f'Projects_and_Volume_{uf}_2010_2021.png'))
    plt.show()

```

```python
# 5) Financiamentos por setor (stacked) - mapeamento simples
sector_map = {
    "Agro": ["agric","agro","agricultura"],
    "Edu_Saude_Admin": ["educ","saúde","saude"],
    "Servicos": ["comércio","comercio","serviço","servico","serviços","servicos","turismo","saneamento"],
    "Industria_Infrastr": ["energia","logística","transporte","infra","indústria","industria","mobilidade","infraestrutura","crédito","financiamento"]
}
def map_sector(s):
    if pd.isna(s): return "Outros"
    s2 = str(s).lower()
    for k,terms in sector_map.items():
        for t in terms:
            if t in s2: return k
    return "Outros"

fin_states['sector_group'] = fin_states['nmSetor'].apply(map_sector)
fin_sector_agg = fin_states.groupby(['UF_sigla','Ano','sector_group'], as_index=False)['VlDolar_num'].sum()

for uf in STATES:
    pivot = fin_sector_agg[fin_sector_agg['UF_sigla']==uf].pivot(index='Ano', columns='sector_group', values='VlDolar_num').fillna(0)/1e6
    if not pivot.empty:
        pivot.plot(kind='bar', stacked=True, figsize=(12,4))
        plt.title(f"Financiamentos por setor ({uf}) — 2010–2021 (US$ milhões)")
        plt.xlabel('Ano'); plt.ylabel('US$ milhões')
        plt.savefig(os.path.join(out_dir,f'Fin_by_sector_{uf}_2010_2021.png')); plt.show()

print("All charts saved in:", out_dir)
```
