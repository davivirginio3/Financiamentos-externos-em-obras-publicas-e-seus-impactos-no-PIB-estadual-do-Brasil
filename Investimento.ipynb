```python
# -*- coding: utf-8 -*-
"""
Análises / gráficos de insights entre PIB estadual e financiamentos (SP e RJ).
Assume que você já leu os arquivos e tem os DataFrames:
- pib_agg: ['UF_sigla','Ano','PIB_valor','VAB_Agro','VAB_Ind','VAB_Serv','VAB_AdminSaude']
- fin_agg: ['UF_sigla','Ano','VlDolar_num']
- merged: merge de pib_agg + fin_agg (colunas adicionais possíveis)
- fin_states: dados individuais de projetos (com colunas 'UF_sigla','Ano','VlDolar_num','nmSetor', ...)
- fin_sector_agg: ['UF_sigla','Ano','sector_group','VlDolar_num']

Se você ainda não tem df_cambio e df_ipca (taxa média anual e índice IPCA),
pode preencher manualmente ou baixar de fontes oficiais. Exemplo assume que
você tem df_cambio e df_ipca com colunas: ['Ano','cambio_RpUS'] e ['Ano','ipca_idx'] (base 2021=100).
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.signal import correlate
from sklearn.cluster import KMeans

# ---------- CONFIGURAÇÕES ----------
UFs = ["SP","RJ"]
ANO_INI = 2010
ANO_FIM = 2021
# se quiser log-transform para regressões/plots:
USE_LOG_WHEN_PLOTTING = True

# ---------- Função: conversão US$->R$ e deflação para base 2021 ----------
def convert_and_deflate(merged_df, df_cambio=None, df_ipca=None, base_year=2021):
    """
    merged_df deve conter colunas: 'Ano', 'VlDolar_num' (US$), 'PIB_valor' (R$)
    df_cambio: ['Ano','cambio_RpUS'] -> média anual R$ por US$
    df_ipca: ['Ano','ipca_idx'] -> índice com base já normalizado? pode ser qualquer base
    base_year: ano para o qual normalizamos (ex.: 2021)
    Retorna merged_df com colunas novas:
      - fin_R$ (convertido)
      - fin_R$_real{base_year} (deflacionado)
      - PIB_real{base_year}
    """
    m = merged_df.copy()
    if df_cambio is None or df_ipca is None:
        print("Aviso: df_cambio ou df_ipca não fornecidos - pulando conversão (remova esta mensagem provendo as séries).")
        return m

    # merge
    m = m.merge(df_cambio[['Ano','cambio_RpUS']], on='Ano', how='left')
    m = m.merge(df_ipca[['Ano','ipca_idx']], on='Ano', how='left')

    # normalizar ipca para base_year = 100
    ipca_base = float(m.loc[m['Ano']==base_year, 'ipca_idx'].dropna().unique()[0])
    m['ipca_idx_base100'] = m['ipca_idx'] / ipca_base * 100.0

    # conversão e deflação
    m['fin_R$'] = m['VlDolar_num'] * m['cambio_RpUS']            # US$ -> R$
    m[f'fin_R$_real{base_year}'] = m['fin_R$'] * (100.0 / m['ipca_idx_base100'])
    m[f'PIB_R$_real{base_year}'] = m['PIB_valor'] * (100.0 / m['ipca_idx_base100'])

    return m

# ---------- Função: correlação simples e regressão ----------
def corr_and_regression(merged_df, uf, value_fin_col='VlDolar_num', value_pib_col='PIB_valor'):
    df = merged_df[(merged_df['UF_sigla']==uf) & (merged_df['Ano'].between(ANO_INI, ANO_FIM))].dropna(subset=[value_fin_col, value_pib_col])
    if df.empty or len(df) < 2:
        print(f"Dados insuficientes para {uf}")
        return None
    x = df[[value_pib_col]].values.reshape(-1,1)
    y = df[value_fin_col].values.reshape(-1,1)
    # regressão linear
    model = LinearRegression().fit(x, y)
    r2 = model.score(x,y)
    coef = model.coef_[0][0]
    intercept = model.intercept_[0]
    print(f"Regressão {uf}: R²={r2:.3f}, coef={coef:.6f}, intercept={intercept:.3f}")
    # plot
    plt.figure(figsize=(6,6))
    if USE_LOG_WHEN_PLOTTING:
        plt.scatter(np.log1p(df[value_pib_col]), np.log1p(df[value_fin_col]))
        xs = np.linspace(np.log1p(df[value_pib_col]).min(), np.log1p(df[value_pib_col]).max(), 100)
        # linha previsão approximada via regresão no espaço original (apenas ilustrativa)
        plt.xlabel(f"log1p({value_pib_col})")
        plt.ylabel(f"log1p({value_fin_col})")
    else:
        plt.scatter(df[value_pib_col], df[value_fin_col])
        xs = np.linspace(df[value_pib_col].min(), df[value_pib_col].max(), 100)
        plt.xlabel(value_pib_col)
        plt.ylabel(value_fin_col)
        plt.plot(xs, model.predict(xs.reshape(-1,1)), color='C1')
    plt.title(f"PIB vs Financiamentos ({uf}) — R²={r2:.3f}")
    plt.grid(True); plt.show()
    return {'model': model, 'r2': r2}

# ---------- Função: lag / cross-correlation ----------
def cross_correlation_lags(merged_df, uf, max_lag=5, fin_col='VlDolar_num', pib_col='PIB_valor'):
    """
    calcula cross-correlation entre séries (financiamentos) e (PIB) por lag (fin -> pib).
    Retorna correlações para lags 0..max_lag (fin lead pib).
    """
    df = merged_df[(merged_df['UF_sigla']==uf) & (merged_df['Ano'].between(ANO_INI, ANO_FIM))].sort_values('Ano')
    s_fin = df[fin_col].fillna(0).values
    s_pib = df[pib_col].fillna(0).values
    corrs = {}
    for lag in range(0, max_lag+1):
        # shift fin forward by lag (i.e., does financing at year t relate to pib at year t+lag)
        if lag == 0:
            a = s_fin
            b = s_pib
        else:
            a = s_fin[:-lag]
            b = s_pib[lag:]
        if len(a) < 2: 
            corrs[lag] = np.nan
        else:
            corrs[lag] = np.corrcoef(a, b)[0,1]
    print(f"Cross-correlation (fin -> PIB) para {uf}: {corrs}")
    return corrs

# ---------- Função: contar projetos vs volume ----------
def projects_count_vs_volume(fin_states_df, pib_agg_df, uf):
    """
    Cria gráfico comparando número de projetos por ano vs volume (US$)
    e scatter PIB vs número de projetos.
    """
    # contagem por ano
    cnt = fin_states_df[fin_states_df['UF_sigla']==uf].groupby('Ano').size().rename('n_projetos').reset_index()
    vol = fin_states_df[fin_states_df['UF_sigla']==uf].groupby('Ano')['VlDolar_num'].sum().rename('vol_USD').reset_index()
    df = cnt.merge(vol, on='Ano', how='outer').fillna(0).sort_values('Ano')
    pib = pib_agg_df[pib_agg_df['UF_sigla']==uf][['Ano','PIB_valor']].set_index('Ano')
    df = df.set_index('Ano').join(pib).reset_index()
    # plots
    fig, ax1 = plt.subplots(figsize=(10,5))
    ax1.bar(df['Ano']-0.2, df['n_projetos'], width=0.4, label='n projetos', alpha=0.8)
    ax1.set_ylabel('Número de projetos')
    ax2 = ax1.twinx()
    ax2.bar(df['Ano']+0.2, df['vol_USD']/1e6, width=0.4, color='C1', label='Volume (US$ mi)', alpha=0.8)
    ax2.set_ylabel('Volume (US$ milhões)')
    plt.title(f"Nº projetos vs Volume por ano — {uf}")
    ax1.set_xlabel('Ano')
    ax1.legend(loc='upper left'); ax2.legend(loc='upper right')
    plt.show()
    # scatter PIB vs n_projetos
    plt.figure(figsize=(6,5))
    plt.scatter(df['PIB_valor']/1e9, df['n_projetos'])
    plt.xlabel("PIB (R$ bilhões)"); plt.ylabel("Número de projetos")
    plt.title(f"PIB vs número de projetos — {uf}")
    plt.grid(True); plt.show()
    return df

# ---------- Função: comparar setores financiamentos vs VAB ----------
def sector_vs_vab(fin_sector_agg_df, pib_agg_df, uf):
    """
    Plota stacked bars dos setores financiados e linhas do VAB setorial.
    """
    df_fin = fin_sector_agg_df[fin_sector_agg_df['UF_sigla']==uf]
    pivot = df_fin.pivot(index='Ano', columns='sector_group', values='VlDolar_num').fillna(0)/1e6  # em US$ milhões
    if not pivot.empty:
        pivot.plot(kind='bar', stacked=True, figsize=(12,5))
        plt.title(f"Financiamentos por setor (US$ milhões) - {uf}")
        plt.xlabel('Ano'); plt.ylabel('US$ milhões'); plt.show()
    # VAB
    df_vab = pib_agg_df[pib_agg_df['UF_sigla']==uf].set_index('Ano')[['VAB_Agro','VAB_Ind','VAB_Serv','VAB_AdminSaude']]/1e9
    plt.figure(figsize=(10,5))
    for c in df_vab.columns:
        if df_vab[c].notna().any(): plt.plot(df_vab.index, df_vab[c], marker='o', label=c)
    plt.title(f"Evolução setorial do VAB (R$ bilhões) - {uf}")
    plt.xlabel('Ano'); plt.ylabel('R$ bilhões'); plt.legend(); plt.grid(True); plt.show()
    return pivot, df_vab

# ---------- Função: financiamento / PIB (intensidade) e per capita ----------
def financing_intensity_and_percapita(merged_df, pop_df=None):
    """
    Calcula:
      - financ/PIB (razão)
      - financiamento per capita (se pop_df fornecida: ['UF_sigla','Ano','pop'])
    """
    df = merged_df.copy()
    df['fin_USD_to_PIB_pct'] = 100.0 * df['VlDolar_num'] * 1.0 / (df['PIB_valor'] + 1e-9)  # parcela (US$/R$) - só indicativo sem conversão
    # se quiser usar fin_R$_real2021, substituir numerator por fin_R$...
    if pop_df is not None:
        df = df.merge(pop_df[['UF_sigla','Ano','pop']], on=['UF_sigla','Ano'], how='left')
        df['fin_USD_percapita'] = df['VlDolar_num'] / df['pop']
    return df

# ---------- Função: Top N projetos (outliers) ----------
def top_projects(fin_states_df, n=10):
    top = fin_states_df.sort_values('VlDolar_num', ascending=False).head(n)
    display_cols = ['nmProjeto','UF_sigla','Ano','nmSetor','VlDolar_num']
    print("Top projects by value (USD):")
    return top[display_cols]

# ---------- Função: Elasticidade (back-of-envelope) ----------
def estimate_elasticity(merged_df, uf, fin_col='VlDolar_num', pib_col='PIB_valor'):
    """
    Estima elasticidade simples: percentual change in PIB / percentual change in financing
    usando regressão nos logs (elasticidade como slope).
    """
    df = merged_df[(merged_df['UF_sigla']==uf) & (merged_df['Ano'].between(ANO_INI, ANO_FIM))].dropna(subset=[fin_col,pib_col]).sort_values('Ano')
    if len(df) < 3:
        print("Dados insuficientes para elasticidade")
        return None
    x = np.log(df[fin_col].replace(0, np.nan)).dropna()
    y = np.log(df.loc[x.index, pib_col])
    # regressão OLS simples
    X = x.values.reshape(-1,1)
    Y = y.values
    model = LinearRegression().fit(X, Y)
    elast = model.coef_[0]
    r2 = model.score(X, Y)
    print(f"Elasticidade estimada (PIB % / Fin %): {elast:.3f}, R²={r2:.3f}")
    # plot
    plt.figure(figsize=(6,5))
    plt.scatter(x, y)
    xs = np.linspace(x.min(), x.max(), 100)
    plt.plot(xs, model.predict(xs.reshape(-1,1)), color='C1')
    plt.xlabel('log(Financiamento)'); plt.ylabel('log(PIB)')
    plt.title(f"Elasticidade (log-log) — {uf}")
    plt.grid(True); plt.show()
    return {'elasticity': elast, 'r2': r2, 'model': model}

# ---------- Função: cluster anos por comportamento (ex.: 2 clusters) ----------
def cluster_years_by_pattern(merged_df, uf, n_clusters=2):
    df = merged_df[(merged_df['UF_sigla']==uf) & (merged_df['Ano'].between(ANO_INI, ANO_FIM))].sort_values('Ano')
    features = df[['PIB_valor','VlDolar_num']].fillna(0).values
    if len(features) < n_clusters:
        print("Poucos anos para clusterização")
        return None
    # padronizar por z-score simples
    feats_std = (features - features.mean(axis=0)) / (features.std(axis=0) + 1e-9)
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(feats_std)
    df['cluster'] = kmeans.labels_
    print(f"Ano -> cluster ({uf}):")
    print(df[['Ano','PIB_valor','VlDolar_num','cluster']])
    return df

# ---------- EXECUTAR ANÁLISES PRINCIPAIS (loop por UF) ----------
def run_all_insights(pib_agg, fin_agg, merged, fin_states, fin_sector_agg, df_cambio=None, df_ipca=None, pop_df=None):
    # se df_cambio e df_ipca fornecidos, convertemos e criamos novas colunas em merged
    if df_cambio is not None and df_ipca is not None:
        merged_conv = convert_and_deflate(merged, df_cambio, df_ipca, base_year=2021)
    else:
        merged_conv = merged.copy()

    results = {}
    for uf in UFs:
        print("\n\n====================")
        print(f"Análises para {uf}")
        print("====================\n")
        # correlação e regressão
        results[uf] = {}
        results[uf]['regression'] = corr_and_regression(merged_conv, uf, value_fin_col=('VlDolar_num' if 'VlDolar_num' in merged_conv.columns else f'fin_R$_real2021'),
                                                       value_pib_col=('PIB_valor' if 'PIB_valor' in merged_conv.columns else f'PIB_R$_real2021'))
        # lag analysis
        results[uf]['lag_corrs'] = cross_correlation_lags(merged_conv, uf, max_lag=4)
        # projects count vs volume
        results[uf]['proj_count_df'] = projects_count_vs_volume(fin_states, pib_agg, uf)
        # sector vs vab
        results[uf]['sector_pivot'], results[uf]['vab_df'] = sector_vs_vab(fin_sector_agg, pib_agg, uf)
        # top projects
        results[uf]['top10'] = top_projects(fin_states, n=10).query("UF_sigla == @uf")
        print(results[uf]['top10'])
        # elasticity
        results[uf]['elasticity'] = estimate_elasticity(merged_conv, uf)
        # cluster anos
        results[uf]['clusters'] = cluster_years_by_pattern(merged_conv, uf, n_clusters=2)
    # intensity / percapita
    results['merged_with_intensity'] = financing_intensity_and_percapita(merged_conv, pop_df)
    return results

# ---------- EXEMPLO DE USO ----------
# Depois de carregar seus DataFrames (pib_agg, fin_agg, merged, fin_states, fin_sector_agg),
# e opcionalmente:
# - df_cambio = pd.read_csv('/path/to/cambio.csv')  # colunas: Ano,cambio_RpUS
# - df_ipca = pd.read_csv('/path/to/ipca.csv')      # colunas: Ano,ipca_idx (qualquer base)
# - pop_df = pd.read_csv('/path/to/pop_estimates.csv')  # colunas: UF_sigla,Ano,pop

# results = run_all_insights(pib_agg, fin_agg, merged, fin_states, fin_sector_agg, df_cambio=df_cambio, df_ipca=df_ipca, pop_df=pop_df)

# Se não quer converte/deflacionar agora, rode sem df_cambio/df_ipca:
# results = run_all_insights(pib_agg, fin_agg, merged, fin_states, fin_sector_agg)

"""
Saídas esperadas em 'results':
- results['SP'], results['RJ'] com chaves: regression, lag_corrs, proj_count_df, sector_pivot, vab_df, top10, elasticity, clusters
- files/plots serão mostrados inline pelo matplotlib.
"""

# Fim do script
```
